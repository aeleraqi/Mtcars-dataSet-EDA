```r
library(corrplot)

correlations <- cor(df)
# create correlation plot
corrplot(correlations, method="circle")
```

```r
pairs(df[c(2,3,4,5)])
```

```r
# Scatterplot of MPG versus Weight using ggplot2
ggplot(df, aes(x = wt, y = mpg)) +
  geom_point(color = "black", alpha = 0.9, shape = 10) +
  geom_smooth(method = "lm", se = FALSE, color = "red") + # Regression line
  labs(title = "Scatterplot of Miles Per Gallon versus Car Weight",
       x = "Car Weight (lb)",
       y = "Miles per Gallon (mpg)") +
  theme_fivethirtyeight()
```

```r
# Scatterplot of CYL versus DISP using ggplot2
ggplot(df, aes(x = cyl, y = disp)) +
  geom_point(color = "black", alpha = 0.9, shape = 10) +
  geom_smooth(method = "lm", se = FALSE, color = "red") + # Regression line
  theme_fivethirtyeight()
```

```r
ggplot(df, aes(x = factor(cyl), y = mpg)) +
  geom_violin() +
  geom_jitter(size = 2, height = 0.05, width = 0.05, color = "darkgray") +
  labs(title = "Distribution of Miles Per Gallon (MPG) for Cars with Different Numbers of Cylinders",
       x = "Number of Cylinders",
       y = "MPG") +
  theme_minimal()
```

```r
ggplot(df, aes(x = mpg, y = hp)) +
  geom_point() +
  facet_grid(rows = vars(cyl), cols = vars(vs)) +
  labs(title = "Scatterplot Matrix: Relationship Between Different Variables",
       x = "Mileage (MPG)",
       y = "Horsepower (HP)") +
  theme_bw()
```

```r
ggplot(df, aes(x = factor(am), y = mpg)) +
  geom_boxplot() +
  labs(title = "Comparison of Miles Per Gallon (MPG) Distributions Across Auto & Manual Transmissions",
       x = "Transmission Type",
       y = "MPG") +
  theme_minimal()
```

```r
ggplot(df, aes(mpg)) +
  geom_density() +
  facet_wrap(~am)
```

```r
boxplot(mpg ~ gear, data = df)
```
